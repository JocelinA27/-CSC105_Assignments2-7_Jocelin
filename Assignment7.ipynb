```python
import pandas as pd
import numpy as np
```

```python
t_df = pd.read_csv("Data/titanic_data.csv")
```

```python
t_df
```

```python
survived_by_class = t_df.groupby(["class"])[["survived"]].sum()
print(survived_by_class)
```

```python
average_fare_town = t_df.groupby(["embark_town"])[["fare"]].mean()
print(average_fare_town)
```

```python
average_fare_sex_town = t_df.groupby(["embark_town", "sex"])[["fare"]].mean()
print(average_fare_sex_town)
```

```python
average_fare_sex_town_age = t_df.groupby(["embark_town", "sex"])[["age", "fare"]].mean()
print(average_fare_sex_town_age)
```
